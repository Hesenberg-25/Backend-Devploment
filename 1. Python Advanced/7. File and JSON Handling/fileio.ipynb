```python
# So traditionally we use use the open command to open a File
# Which contain the Path of the Database if it is present in some other Directory
# When wee Open a file and move our Cursor to some places headd it will perform the reat operation for that file if the File is kept open

f=open('test.txt','w')
# So while opening the file we need to give Mode of Opening along wiht the Path by default it is Readmode
# 'w'  to write
# 'r'  to read
# 'a'  to append
# 'r+' to read and write

print(f.name)
# To get the name of the File

print(f.mode)
# To print the Mode of File Opening

f.close()
# And we need to Manually close the file
```

```python
# So to Overcome the Manual work of Closing the File we use Context Manager
with open('test.txt','r') as f:
    print(f.name)
    print(f.mode)
```

```python
with open('test.txt','r') as f:
    content=f.read()
    print(content)

# When we want to Read content inside the file
    f.seek(0)
    
    content_readlines=f.readlines()
    print(content_readlines)
# This gives us the List of all the Lines in the Files
```

```python
# Sometimes Files are so Big that we may Run out of Memory while Reading them so we use Itration
with open('test.txt','r') as f:
    for lines in f:
        print(lines,end='')
```

```python
with open('test.txt','r') as f:
    content=f.read(25)
    # By giving input of Int inside the read() we cn control how much Character we want to Display
    print(content)
```

```python
with open('test.txt','r') as f:
    size_to_read=10

    f_content=f.read(size_to_read)
    while len(f_content)>0:
        print(f_content,end="*")
        f_content=f.read(size_to_read)
```

```python
with open('test.txt','r') as f:
    size_to_read=10

    contents=f.read(size_to_read)
    print(f.tell())

# f.tell() tells us the current Postion of the Cursor in the Database
```

```python
# Now to write inside a File
# So we can Use to Method ::
# 1. 'w' which Overwrite what so ever content the newly Enter content needs character till
with open('test.txt','w') as f:
    f.write('Testing the File here')

# Like here first we entered 'Testing the File here' then moved Curosor to seek(0)
# And write('R') till will just Overwrite the Space needed by the Newly entered Character 'R'
    f.seek(0)
    f.write('R')

# 2. 'a' which do not concerns with the seek() as it cant overwrite so it just add new Content at End of the File
```

```python
# How can we copy Content from one File and put it in Another

with open('test.txt','r') as rf:
    contentinside=rf.read()

    with open('test2.txt','w') as wf:
        wf.write(contentinside)  
```

```python
# Another way

with open('test.txt','r') as rf:
    with open('test2.txt','w') as wf:
        for line in rf.read():
            wf.write(line)
```
