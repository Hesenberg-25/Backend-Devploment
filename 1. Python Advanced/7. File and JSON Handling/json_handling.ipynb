```python
# JSON stands for JavaScript Object Notation
# It is used to send DATA back and forth between a Server and a Client
# Or between diffrent Server like API
```

```python
import json

people_string = '''
{
    "people":[
        {
            "name": "Test User",
            "phone": "615-555-7164",
            "emails": "TestUser@gmail.com",
            "has_lic": false
        },
        {
            "name": "Jane Doe",
            "phone": "560-555-1234",
            "emails": null,
            "has_lic": true
        }
    ]
}
'''
```

```python
data=json.loads(people_string)
# json.loads helps in loading a string and json.load help in loading DataBase

print(type(data))
print(data)
print(type(data['people']))

# So we can say that there is list with name 'person'
# Whose Elemnts are specific to one Entity which give in form of key:value pair for that Entity

# And as the internal info is in form of Dict we can ascess them like the Key:Value pair of Python
```

```python
for person in data['people']:
    print(person)
    print(person['name'])

# If you want to access the Specific Entity from the list 'people' we use the same method as the Index of List
print(data["people"][0]["name"])
print(data["people"][1]["emails"])
```

```python
# We should also know how can we put rrhe Python data into JSON again
# for person in data['people']:
#     del person['phone']

new_data=json.dumps(data,indent=2)
print(new_data)

# indent help us to design the Inside data in a Visually Good way
```

```python
# First we need to Open the json file in read Mode then Load it as Data
with open('states.json','r') as f:
    data=json.load(f)

for states in data['states']:
    print(states['name'],states['initials'])

# We can print the Values for the keys we have entered
```

```python
for states in data['states']:
    del states['state_codes']

with open('new_states.json','w') as wf:
    json.dump(data,wf,indent=2) 
```
