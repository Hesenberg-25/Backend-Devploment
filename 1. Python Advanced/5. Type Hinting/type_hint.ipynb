```python
from typing import NewType

# Again sometimes using Type Alias can  cause Error while using them
# More than 1 Type Alias with same return type may cause unwanted Confusion

# => type RGB=tuple[int,int,int]
# => type HSL=tuple[int,int,int]

# These two have same return type then if we passout wrong input it will not show ua any Error
RGB=NewType("RGB",tuple[int,int,int])
HSL=NewType("HSL",tuple[int,int,int])

type User=dict[str,str|int|RGB|HSL|None]

def create_user(firstname:str,lastname:str,age:int|None=None,fav_color:RGB|HSL|None=None)->User:
    email=f"{firstname}{lastname}@gmail.com"

    return{
        "FirstName": firstname,
        "Lastname":lastname,
        "Age":age,
        "Email":email,
        "FavColor":fav_color
    }

user1=create_user("Test","User",18,fav_color=RGB((123,321,159)))
user2=create_user("Test","User",18,fav_color=HSL((1,2,3)))

# So here we need to Specify the type of the Return Type to be Sure 

print(user1)
print(user2)
```

```python
# Now we know that TypedHint is a Static Method to HintChecking
# Which can be sometimes bypassed using External API input which just Return Dict
# So over that we just you Dataclass so that we Return in from class

from dataclasses import dataclass 

@dataclass
class User:
    FirstName:str
    LastName:str
    Age:int|None
    Email:str
    FavColor:RGB|HSL|None

def create_user(firstname:str,lastname:str,age:int|None=None,fav_color:RGB|HSL|None=None)->User:
    email=f"{firstname}{lastname}@gmail.com"

    str_age=str(age)
    return User(
        FirstName= firstname,
        LastName=lastname,
        Age=age,
        Email=email,
        FavColor=fav_color
    )


user1=create_user("Test","User",18,fav_color=RGB((123,321,159)))
user2=create_user("Test","User",18,fav_color=HSL((1,2,3)))

print(user1)
print(user2)
```
