```python
"""
A Logger is a dedicated runtime object in a software application that serves as the entry point for capturing and categorizing execution events, milestones, and errors.
It is responsible for exposing severity-level methods like .info() or .error(), performing the initial filtering of messages against a configured threshold level, and routing passing records to handlers for storage or display.
"""
import logging
import employee

def addition(x:int,y:int):
    return x+y
def subtraction(x:int,y:int):
    return x-y
def multiplication(x:int,y:int):
    return x*y
def divide(x:int,y:int):
    return x/y

logging.basicConfig(filename='sample.txt',level=logging.DEBUG)

num1=20
num2=10

add_res=addition(num1,num2)
logging.info("Add: {} + {} = {}".format(num1,num2,add_res))

sub_res=subtraction(num1,num2)
logging.info("Sub: {} - {} = {}".format(num1,num2,sub_res))

mul_res=multiplication(num1,num2)
logging.debug("Mul: {} * {} = {}".format(num1,num2,mul_res))

div_res=divide(num1,num2)
logging.info("Div: {} / {} = {}".format(num1,num2,div_res))
"""
1. So the reason we are not to log the Operation into sample.txt because is because we have already opened in level of info
2. And the root Logger is Singleton it measn it can be changed just once
3. And now if you change the level from debug to info the program will Automatically log in the Intial log File

"""
```
