```python
# In python Class called Exception which is in-build holds all the needed Exceptions
# From which all the Errors can be Inheritedd to Raise Custom Exception 
```

```python
# Like Suppose you want to Take Email from User but you also need to be Sure that the Email is not Already Used
class EmailAlreadyExistError(Exception):
    def __init__(self, email:str):
        self.email=email
        self.message=f"{email} already Registered."
        
        super().__init__(self.message)

ExistingUser=["testuser@gmail.com"]

def generateUser(email:str):
    if email in ExistingUser:
        raise EmailAlreadyExistError(email)
    print(f"User with Email {email} Generate Sucessfully")

user1=generateUser("testuser@gmail.com")
print(user1)

# If we try to Enter anyother Valid Email This Error will not PopUp
```

```python
class AgeLimitError(Exception):
    def __init__(self,age:int):
        self.age=age
        self.message="Entered Age is not Valid"

        super().__init__(self.message)

def enterAge(age:int):
    if age<18:
        raise AgeLimitError(age)
    print("Entered Age stored Sucessfully")

user1=enterAge(7)
```
