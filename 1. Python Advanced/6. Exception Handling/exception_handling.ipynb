```python
# Suppose we create a File name exp.txt and try to open it
# But we may soemtimes cause Typo while write so to Handle this Isuue we use Try Execept
# Like Suppose we have created a Database with name file.txt

try:
    open('exps.txt')
except:
    print('File not Found')
```

```python
# But sometimes it may also happen like we have also Coded some Code in the same Try block which can cause Error
# But our file open Code is Correct then too it will show the 'File not Found' Error
try:
    f=open('exp.txt','r')
    var=bad_var
except FileNotFoundError:
    print('File not Found')

# So here we Specify using the except 'FileNotFoundError' to display the message only wehen the File is not Found
# Which will here give a Name Error as the except block only runs when file is not found
```

```python
# So we do overcome this by added some Additonal other Except Statement

try:
    open('exp.txt')
    var=var_bad
except FileNotFoundError:
    print('File not Found')
except Exception:
    print('Something went Wrong')

# Remember the except statement which are more Specific should be at Top 
# And General ones at Last this is because the Multiple Exception block work like Ladder
# Means the ones at top gets Executed first than below ones so there should be order
```

```python
# If we want to Print the Same Error as we get we use 'as'
try:
    open('exp.txt')
    var=var_bad
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)

# This we Display the Machine Level Error Format
```

```python
# Now the work of else block is to Excecute the Code which is Needed to Executed when the Try Block do not Throw any Error
try:
    f=open('exp.txt')
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close

# We can also Execute the 'else' block code in 'try' Block
# But we need to be More Specific about what we are Trying to check here which my Ease our DeBugging
```

```python
# Now the use case of Finally is that No matter what Happen with the Try Except Block this Block is Must to be Executed no Matter what

try:
    f=open('exp2.txt')
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close
finally:
    print('DataBase Closed')
```

```python
# You can also Raise some Custom Exception Whernever Needed using Syntax raise {Exception Name}

try:
    f=open('exp.txt')
    if f.name=='exp.txt':
        raise Exception
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print('Error')
else:
    print(f.read())
    f.close
finally:
    print('DataBase Closed')
```
